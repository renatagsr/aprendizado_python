#Lidando com Dados Ausentes

Na análise de dados é bem comum que encontrar uma base de dados que esteja faltando algum dado, então saber manipular isso é extremamente importante para que alcance o objetivo desejado.

O primeiro conjunto de dados que será tratado nesse notebook são os dados de crimes no Rio de Janeiro, esses dados são mensais e vão desde Janeiro de 1991. Alguns crimes, como o roubo de celular, começaram a serem notificados alguns anos depois, então terá dados faltantes.

Quem lida com dados passa a maior parte do tempo no tratamento de dados, análise exploratória, etc, assim se percebe a importância do conhecimento de saber tratar os dados adequadamente.

## Como lidar com dados ausentes

Há diversas formas de lidar com dados ausentes, isso se deve ao fato de que cada conjunto de dados irá apresentar dados ausentes de uma forma ou por um motivo.

No conjunto de dados sobre crimes do Rio de Janeiro, por exemplo, ao fazer uma análise percebe-se que o motivo de haver dados faltantes é porque alguns crimes começaram ser registrados depois de 1991.

O ".isnull()" mostra onde há dados faltantes no conjunto de dados que estão sendo tratados. Só que melhor que mostrar da forma que está sendo monstrando abaixo, é melhor fazer uma soma de quantos valores ausentes tem em cada uma das colunas, pois assim irá gerar uma resposta mais relevante.

```
df.isnull()
```

```
df.isnull().sum()
```

Proporção de dados faltantes em relação ao total de linhas em cada coluna.

```
(df.isnull().sum()/df.shape[0]) * 100
```

```
#dados ausentes por ano
df.set_index('vano').isnull().sum(level=0)
```

Essa resposta gerada comprova o que já era imaginado, os dados faltantes estão nos primeiros anos, ou seja, alguns crimes começaram a ser registrados depois de ter passado alguns anos.

## Tratando dados 

Para o conjunto de dados de crimes no Rio de Janeiro o ideal para tratar os dados é considerar a partir de onde há os dados registrados. Para outros conjuntos de dados poderia se excluir, caso sejam poucos dados e eles estejam registrados de forma aleatória ou então pode preencher os dados faltantes seja com a média, mediana, zeros. Recomenda-se substituir os dados faltantes com a média quando a distribuição é normal, já para dados com a distruibuição mais enviesada é melhor substituir com a mediana, já que a média é muito mais sensivel a outliers. Substituir a ausência com algum valor pode ser interessante, por exemplo, em dados númericos pode substituir ela por zeros e em dados categóricos substituir com "desconhecido". É importante lembrar que em dados estatísticos não pode substituir os valores ausentes com zeros. 

```
df['roubo_celular']
```

```
media_roubo_celular = df['roubo_celular'].mean()
df['roubo_celular'] = df['roubo_celular'].fillna(media_roubo_celular)
```

## Segundo conjunto de dados

Esses dados foram retirados do Inside Airbnb, ele possui apenas algumas linhas e colunas com dados ausentes.

```
df_nyc = pd.read_csv('https://raw.githubusercontent.com/rafaelnduarte/eds_outliers/master/nyc.csv', index_col = 0)

#Dimensões do conjunto de dados
print('Dimensões do Dataset',
    '\nVariáveis: ',df_nyc.shape[1], "\n"
      'Entradas: ', df_nyc.shape[0])

#Visualizando o dataset
df_nyc.head()
```

```
df_nyc.isnull().sum().sort_values(ascending = False)
```

As colunas "reviews_per_month" e "last_review" por serem colunas que tem muitos dados faltantes e por serem escritas (cada pessoa da uma resposta diferente), será assumido essas colunas não seriam utilizadas e, portanto, poderão ser excluídas sem prejuízos.

As demais colunas possuem poucas informações ou nenhuma informação faltante, para as coluns que possuem poucas informações faltantes será feita a retirada desse valores.

```
#Excluindo coluna com dados faltantes
df_nyc = df_nyc.drop(columns = ['reviews_per_month', 'last_review'])

#Excluindo entradas com dados faltantes
df_nyc = df_nyc.dropna(axis = 0)
```
