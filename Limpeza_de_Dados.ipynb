Com o resultado obtido, percebe-se que a coluna "price" é do tipo objeto e não do tipo float conforme é desejável visto que são preços. Com isso, é necessário qu converta as strings em floats.

```
valor = '$244.00'
print(type(valor))
valor = float(valor.replace('$',''))
print(valor)
print(type(valor))
```

## Funções Lambda

Resumidamente, funções lambdas são aquelas funções que é possível colocar ela em apenas uma linha. Elas são funções anônimas.

A sintaxe utilizada para criar uma função lambda é bem simples: 

> lambda argumento_1, argumento_2 …aurgumento_n: expressão

Essas funções são normalmente utilizadas quando é necessário utilizar uma função dentro de outra função, quando se deseja algo mais simplificado ou quando não será necessário a repetição da função mais de uma vez.

Um ponto importante é que o *return* das funções lambdas já estão implicítos nela, por isso que não precisa declarar ele.



O "x" na função lambda a cada momento ele irá assumir um resultado até que passe por todas as linhas do dataframe.

```
df['price'].apply(lambda x: float(x.replace('$', '')))
```

Com esse erro descobre-se que foi previsto apenas um tipo de situação para transformar a string em float, esse erro mostra que os números também possuem vírgulas para mostrar que está na casa de milhar, portanto é necessário fazer a retirada dessa vírgula.

```
df['price'].apply(lambda x: float(x.replace(',', '').replace('$', '')))
```
